# Değişken Standardizasyonu(Veri Standardizasyonu)

```python
import numpy as np
import pandas as pd
```

```python
V1= np.array([1,3,6,5,7])
V2= np.array([7,7,5,8,12])
V3= np.array([6,12,5,6,14])
df = pd.DataFrame(
     {'V1': V1,
      'V2': V2,
      'V3': V3})
```

```python
df =df.astype(float)
```

```python
df
```

### Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

```python
df.mean()
```

### Normalizasyon

```python
preprocessing.normalize(df)
```

### Min-Max Dönüşümü

```python
scaler = preprocessing.MinMaxScaler(feature_range= (25,200))
```

```python
scaler.fit_transform(df)
```

## Değişken Dönüşümleri
* not Değişken dönüşümleri > Standardizasyon , yani her bir Standardizasyon bir değişken dönüşümüdür zaten ! 

```python
import seaborn as sns
df= sns.load_dataset('tips')
df.head()
```

#### 0-1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbe= LabelEncoder()
```

```python
lbe.fit_transform(df.sex)
```

```python
df['new_sex'] = lbe.fit_transform(df.sex)
```

```python
df
```

### "1 ve Diğerleri(0)" dönüşümü

```python
df.head()
```

```python
df['new_day'] = np.where(df.day.str.contains('Sun'), 1 , 0)
```

```python
df
```

### Çok sınıflı Dönüşüm

```python
from sklearn.preprocessing import LabelEncoder
lbe= LabelEncoder()
```

```python
lbe.fit_transform(df.day)
```
