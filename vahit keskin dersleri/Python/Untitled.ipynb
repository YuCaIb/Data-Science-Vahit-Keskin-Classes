```python
t.pencolor("white")

def curve():
    for i in range(200):
        t.rt(1)
        t.fd(1)
def heart():
    t.fillcolor('red')
    t.begin_fill()
    t.lt(40)
    t.fd(113)
    curve()
    t.lt(120)
    curve()
    t.fd(112)
    t.end_fill()
heart()
t.ht()

def write(message,pos):
    x,y = pos
    t.penup()
    t.goto(x,y)
    t.color('white')
    style= ('stencil std', 18, 'italic')
    t.write(message,font=style)
    
write('E',(-180,95))
write('V',(-160,95))
write('D',(-140,95))
write('E',(-100,95))
write('K',(-90,95))
write('A',(-70,95))
write('L',(-50,95))
write('D',(-30,95))
write('I',(-10,95))
write('N',(-60,95))

write(' ',(-50,95))

write(':',(-40,95))
write('(',(-30,95))
```
