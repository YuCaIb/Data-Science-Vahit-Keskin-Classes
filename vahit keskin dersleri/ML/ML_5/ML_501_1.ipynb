# K- Means (bir kümeleme yöntemi)
* Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır.
* Kümeleme yöntemlerinin amacı benzerlik matrislerini kullanarak, gözlemleri ya da değişkenleri kümelemeye çalışmaktır.
* Oluşturulmaya çalışılan kümelerin kendi içlerinde homojen, birbirlerine göre heterojen olması beklenir.
---
* K means
1. Küme sayısı belirlenir.
2. Rastgele k merkez seçilir.
3. Her gözlem için k merkezlere uzaklıklar hesaplanır.
4. Her gözlem en yakın olduğu merkeze yani kümeye atanır.
5. Atama işlemlerinden sonra oluşan kümeler için tekrar merkez hesaplamaları yapılır.
6. Bu işlem belirlenen bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının toplamını(total within culester variaton) minumum olduğu durumdaki gözlemlerin kümelenme yapısı nihai kümelenme olarak seçilir.
-----
* küme içi benzerliğin maksiumum olmasını, Kümeler arası benzerliğin minumum olmasını istiyoruz ! amaç bu.

Optimum küme saysının belirlenmesi (birazcık tune gibi)

Elbow yöntemi

```python
df
```

```python
ssd = []
K = range(1,30)
for k in K:
    kmeans= KMeans(n_clusters=k).fit(df)
    ssd.append(kmeans.inertia_)
    
```

```python
plt.plot(K,ssd,'bx-')
plt.xlabel('farklı k değerlerine karşılık uzaklık artık toplamları')
plt.title('optimum küme sayısı için elbow yöntemi')
```
