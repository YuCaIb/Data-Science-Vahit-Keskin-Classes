```python
(sns
 .FacetGrid(df,
               hue='clarity',
              height=5,
              xlim= (0,20000),
              )
 .map(sns.kdeplot,'price',fill=True).add_legend());
# clarity: temizliÄŸi, berraklÄ±ÄŸÄ± (I1(worst), SI2,SI1,VS2,VVS2,VVS1,IF(best))
```

# Korelasyon Grafikleri
### Scatter plot

```python
import seaborn as sns
import pandas as pd
tips= sns.load_dataset('tips')
df=tips.copy()
df.head()
```

```python
df["time"] = df["time"].replace("Dinner", 0)
df["time"] = df["time"].replace("Lunch", 1)

df["sex"] = df["sex"].replace("Female", 1)
df["sex"] = df["sex"].replace("Male", 0)

df["smoker"] = df["smoker"].replace("Yes", 1)
df["smoker"] = df["smoker"].replace("No", 0)
```

```python
df.time=df.time.astype('int')
df["sex"] = df["sex"].astype("int")
df.smoker = df.smoker.astype('int')
```

```python
df.info()
```

```python
df.time.value_counts()
```

```python
sns.scatterplot(x= df.total_bill, y = df.tip);
```
